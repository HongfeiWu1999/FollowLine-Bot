```python
import imageio.v3 as iio
import numpy as np
import image_manager as imm
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D as ax3d
```

```python
# Leo las imagenes de entrenamiento
imNp = iio.imread('images/original.png')
markImg = iio.imread('images/edited.png')
```

```python
def getImageMask(img, color):
    return np.all(img == color, axis=-1)
```

```python
red = [0,0,255]
green = [0,255,0]
blue = [255,0,0]
# Almaceno los valores RGB de 'imNp' que fuero marcados en 'markImg' en rojo verde y azul
data_marca = imNp[getImageMask(markImg, red)]# datos marcados en rojo
data_fondo = imNp[getImageMask(markImg, green)]# datos marcados en verde
data_linea = imNp[getImageMask(markImg, blue)]# datos marcados en azul
```

```python
# Pinto los datos RGB
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(data_marca[:,0],data_marca[:,1],data_marca[:,2],'r.',label='marca')
ax.plot(data_fondo[:,0],data_fondo[:,1],data_fondo[:,2],'g.',label='fondo')
ax.plot(data_linea[:,0],data_linea[:,1],data_linea[:,2],'b.',label='linea')
ax.set_xlabel('R Label')
ax.set_ylabel('G Label')
ax.set_zlabel('B Label')
ax.legend()
plt.title('Espacio RGB')
```

```python
# Ahora haglo lo mismo con los datos normalizados segun la 'normalizacion rgb'


imrgbn = imm.normalizeImage(imNp)# Divido el valor RGB de cada pixel por la suma de los tres valores
imNp=imrgbn

# Almaceno los valores RGB de 'imrgbn' que fueron marcados en 'markImg' en rojo verde y azul
data_marca_n = imNp[getImageMask(markImg, red)]# datos marcados en rojo
data_fondo_n = imNp[getImageMask(markImg, green)]# datos marcados en verde
data_linea_n = imNp[getImageMask(markImg, blue)]# datos marcados en azul
```

```python
plt.figure()
plt.plot(data_marca_n[:,0],data_marca_n[:,1],'r.',label='marca')
plt.plot(data_fondo_n[:,0],data_fondo_n[:,1],'g.',label='fondo')
plt.plot(data_linea_n[:,0],data_linea_n[:,1],'b.',label='linea')
plt.title('Espacio RGB normalizado')
```
